```python
import pandas as pd
```

```python
artworks = pd.read_csv('artworks.csv')
```

```python
top_artists = artworks.groupby('artist').agg('count').sort_values(by='url', ascending=False)[:50]
```

```python
top_artists['url'].sum()
```

```python
artworks_reduced = artworks[artworks['artist'].isin(list(top_artists.index))]
```

```python
artworks_reduced.to_csv('artworks_reduced.csv')
```
